```python
import pandas as pd

# Load dataset
dataset = "D:\logistic_regression_100.csv"
df = pd.read_csv(dataset,usecols=['age', 'salary', 'purchased'])
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['purchased'], df['salary'], color='blue', label='Data points')
plt.xlabel("age")
plt.ylabel("salary")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['age'], df['purchased'], color='blue', label='Data points')
plt.xlabel("age")
plt.ylabel("purchased")
```

```python
X=df[['age', 'salary']]
y=df['purchased']

print(X.head())
print(y.head()) 
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, train_size=0.70, random_state=42)


print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape) 
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)    
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(results.head())
model.score(X_test, y_test)
```

```python
plt.plot(y_test.values, label="Actual", marker='o')
plt.plot(y_pred, label="Predicted", marker='o')
plt.xlabel("Sample Index")
plt.ylabel("Target Value")
plt.title("Actual vs Predicted (Line Plot)")
plt.legend()
plt.show()
```
